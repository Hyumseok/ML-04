```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/drive/MyDrive/Machine Learning/diamonds.csv')
data.head()
```

```python
data.info()
```

#Feature Statistic

```python
# Menampilkan Feauture Statistic
print(data.describe)
```

```python
# Memisahkan fitur yang bersifat numerik dalam satu variabel
feature_numeric = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']
```

#Box Plot

```python
# ChatGPT: Cara menampilkan Box Plot untuk fitur numerik
for feature in feature_numeric:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=data[feature])
    plt.title(f'Boxplot of {feature}')
    plt.show()
```

#Scatter Plot

```python
# ChatGPT: Cara menampilkan Scatter Plot untuk fitur numerik
sns.pairplot(data[feature_numeric])
plt.show()
```

#Violin Plot

```python
# ChatGPT: Cara menampilkan Violin Plot untuk fitur numerik
for feature in feature_numeric:
    plt.figure(figsize=(8, 6))
    sns.violinplot(x=data[feature], inner='quartile')
    plt.title(f'Violin Plot of {feature}')
    plt.xlabel(feature)
    plt.show()
```

#Distribution Plot

```python
# ChatGPT: Cara menampilkan Distribution Plot untuk fitur numerik
for feature in feature_numeric:
    plt.figure(figsize=(8, 6))
    sns.histplot(data[feature], kde=True, bins=30)
    plt.title(f'Distribution Plot of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

#Heat Map

```python
# ChatGPT: Cara menampilkan Heat Map untuk fitur numerik
# Menghitung korelasi pada fitur numerik
corr = data[feature_numeric].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap of Numerical Features Correlation')
plt.show()
```
